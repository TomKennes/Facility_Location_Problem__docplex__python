```python
import pandas as pd
from math import sin, cos, sqrt, atan2, radians
```

```python
distance_DF = pd.read_csv("worldcities.csv").iloc[0:500,:]
```

```python
def coordinates_to_distance(xlat, xlon, ylat, ylon):
    # approximate radius of earth in km
    R = 6373.0
    
    lat1 = radians(xlat)
    lon1 = radians(xlon)
    lat2 = radians(ylat)
    lon2 = radians(ylon)

    dlon = lon2 - lon1
    dlat = lat2 - lat1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    
    return distance
```
